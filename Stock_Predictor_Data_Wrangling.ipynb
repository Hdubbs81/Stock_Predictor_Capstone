```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import os
import csv
from tqdm.notebook import tqdm
from datetime import datetime, timezone
```

```python
# Create tqdm instance to display progress bar
folder_path = '/Users/heatheradler/Documents/GitHub/Springboard/Springboard_Projects/Capstone/archive/Data'
csv_files = os.listdir(folder_path)
progress_bar = tqdm(total=len(csv_files), desc='Processing CSV files')
```

```python
def count_lines_csv_in_folder(folder_path):
    line_counts = {}
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            line_count = count_lines_csv(file_path)
            line_counts[filename] = line_count
    return line_counts

def count_lines_csv(file_path):
    line_count = 0
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line_count += 1
    return line_count

folder_path = '/Users/heatheradler/Documents/GitHub/Springboard/Springboard_Projects/Capstone/archive/Data/Test'
line_counts = count_lines_csv_in_folder(folder_path)
for filename, count in line_counts.items():
    print(f"Number of lines in {filename}: {count}")
    
# Sum all the counts
total_count = sum(line_counts.values())
print("Total number of line items in all CSV files:", total_count)
```

```python
# Set the directory where your CSV files are located
folder_path = '/Users/heatheradler/Documents/GitHub/Springboard/Springboard_Projects/Capstone/archive/Data/StockHistory'

# Initialize an empty list to store dataframes
dfs = []

# Iterate through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Extract the stock symbol from the file name
        stock_symbol = os.path.splitext(file_name)[0]
        
        # Load the CSV file into a dataframe
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        
        # Add a new column 'stock_symbol' with the stock symbol
        df['stock_symbol'] = stock_symbol
        
        # Append the dataframe to the list
        dfs.append(df)
        progress_bar.update(1)

# Concatenate all dataframes together
concatenated_df = pd.concat(dfs, ignore_index=True)

# save concatenated dataframe
concatenated_df.to_csv('/Users/heatheradler/Documents/GitHub/Springboard/Springboard_Projects/Capstone/archive/Concated_Dataframe.csv')
```
